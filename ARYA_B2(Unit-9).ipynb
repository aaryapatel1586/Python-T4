```python
# overloading
def sum_num(*args):
    result=0
    for num in args:
        result+=num
    print("sum:",result)
print("Similar to method overloading concept")
print("Single argument->",end="")
sum_num(10)
print("Two argument->",end="")
sum_num(10,20)
print("multiple argument->",end="")
sum_num(10,20,30,40,50)
```

```python
# comparison operator
class A:
    def __init__(self,a):
        self.a=a
    def __gt__(self,other):
        if self.a>other.a:
            return True
        else:
            return False
obj1=A(8)
obj2=A(7)
if obj1>obj2:
    print("Obj1 is greater")
else:
    print("Obj2 is greater")
```

```python
# program-1
class A:
    def __init__(self,a):
        self.a=a
    def __mul__(self,other):
        return self.a*other.a
    def __floordiv__(self,other):
        return self.a//other.a
    def __lt__(self,other):
        if self.a<other.a:
            return True
        else:
            return False
    def __eq__(self,other):
        if self.a==other.a:
            return True
        else:
            return False
        
obj1=A(9)
obj2=A(1)
print(obj1*obj2)
print(obj1//obj2)
print(obj1.__mul__(obj2))
print(obj1.__floordiv__(obj2))
if obj1<obj2:
    print("Obj1 is less than obj2")
else:
    print("Obj1 is greater")
if obj1==obj2:
    print("Obj1 is equal to obj2")
else:
    print("Obj1 is not equal to obj2")    
    
```
