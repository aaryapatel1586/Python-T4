```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':10}
plt.plot(x,y)
plt.xlabel("Average pluse",fontdict=font2)
plt.ylabel("Calorie burnage",fontdict=font2)
plt.title("Sports Watch Data",fontdict=font1)
plt.grid(color='green',linestyle=':',linewidth=1) 
plt.show()
```

### fill between two lines 

```python
import matplotlib.pyplot as plt
import numpy as np
time=np.arange(12)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expense=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time,income,color='green')
plt.plot(time,expense,color='red')
plt.fill_between(time,income,expense,where=(income>expense),color='green',alpha=0.25,label='positive',interpolate=True)
plt.fill_between(time,income,expense,where=(income<=expense),color='red',alpha=0.25,label='negative',interpolate=True)
plt.xlabel("time")
plt.ylabel("income/expense")
plt.title("DATA")
plt.legend()
plt.show()
```

## Stacked Area chart 

```python
import matplotlib.pyplot as plt
import numpy as np
x=range(1,6)
y1=[1,4,6,8,4]
y2=[2,2,7,10,12]
y3=[2,8,5,10,6]
plt.stackplot(x,y1,y2,y3,labels=['A','B','C'],colors=['blue','yellow','red'])
plt.legend(loc='best')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=range(1,6)
y1=[2,2,2,2,2]
y2=[2,2,2,2,2]
y3=[2,2,2,2,2]
plt.stackplot(x,y1,y2,y3,labels=['A','B','C'],colors=['yellow','red','black'])
plt.title("Germany")
plt.axis("off")
plt.show()
```

# pie chart

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
plt.pie(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
mylabels=['Apple','banana','cherry','dates']
plt.pie(y,labels=mylabels)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
plt.figure(figsize=(10,10))
mylabels=['Apple','banana','cherry','dates']
myexplode=[0.2,0,0,0.2]
mycolors=['red','yellow','pink','gray']
plt.pie(y,labels=mylabels,explode=myexplode,colors=mycolors,radius=0.5,autopct="%1.1f%%",shadow=True)
plt.legend(title='four fruits')
plt.title("pie chart")
plt.show()
```
